```python
import pandas as pd

# Scarica il dataset in formato CSV
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
# Carica il dataset direttamente da Internet
df = pd.read_csv(url)

# Si puÃ² avere il dataset, inoltre, specificando il percorso del proprio file CSV
# file_path = 'percorso_del_tuo_file.csv'
# Da sostituire con il percorso effettivo
```

```python
# Carica il dataset
# df = pd.read_csv(file_path)

# Verifica le dimensioni del dataset
print(f"Dimensioni del dataset: {df.shape}")

# Verifica le diciture presenti nell'intestazione
# mostrando le label delle colonne del dataframe
print(f"\nIntestazione del dataset: {list(df.columns)}")

# o visualizzando anche le prime n righe del dataframe, per esaminare il formato dei dati
print("\nIntestazione del dataset:\n", df.head())

# Filtra le location che non appartengono a nessun continente
df_continents = df.dropna(subset=['continent'])
```

```python
# Raggruppa per paese (location) e continente ottenendo l'ultimo dato disponibile per total_cases e total_vaccinations
# Utilizzo reset_index() per ottenere un DataFrame piÃ¹ malleabile (per le operazioni di selezione, analisi e filtro)
# e piÃ¹ simile a quello di partenza dopo un'operazione di raggruppamento.
df_last_total_cases = df.groupby(['continent', 'location'])['total_cases'].last().reset_index()
df_last_total_vaccinations = df.groupby(['continent', 'location'])['total_vaccinations'].last().reset_index()

# Somma cumulativa dei casi e delle vaccinazioni per ogni continente
df_total_cases_continent = df_last_total_cases.groupby('continent')['total_cases'].sum().reset_index()
df_total_vaccinations_continent = df_last_total_vaccinations.groupby('continent')['total_vaccinations'].sum().reset_index()

# Stampa il numero totale di casi e di vaccinazioni per ogni continente
print("\nNumero totale di casi per continente:\n", df_total_cases_continent)
print("\nNumero totale di vaccinazioni per continente:\n", df_total_vaccinations_continent)

# Stampa il numero totale di casi e vaccinazioni per ogni continente, in base alla locazione
print("\nNumero totale di casi per continente, in base alla locazione:\n", df_last_total_cases)
print("\nNumero totale di vaccinazioni per continente, in base alla locazione:\n", df_last_total_vaccinations)

```

```python
# Calcola e confronta i descrittori statistici tra due continenti e il mondo per il numero di casi covid
def confronto_descrittori_statistici_casi_continenti_mondo(dataset, continente1, continente2):
   subset = dataset[(dataset["continent"] == continente1) | (dataset["continent"] == continente2)]

   total_cases_sum = dataset[dataset["continent"].notna()]["total_cases"].sum()
   perc_cases = round((subset["total_cases"].sum() / total_cases_sum) * 100, 6) # Arrotonda la percentuale dopo la virgola
   
   # Calcola il minimo, il massimo, la media e la somma dei valori in total_cases, rispetto al mondo, formattati con i separatori per le cifre
   min = "{:,.2f}".format(subset["total_cases"].min())
   max = "{:,.2f}".format(subset["total_cases"].max())
   mean = "{:,.2f}".format(subset["total_cases"].mean())
   sum = "{:,.2f}".format(subset["total_cases"].sum())

   descr_statistici = {
      "Minimo": min,
      "Massimo": max,
      "Media": mean,
      "Somma": sum,
      "Percentuale (%)": perc_cases
   }

   return descr_statistici

```

```python
continente1_casi = "Africa"
continente2_casi = "Europe"
desc_stat_casi_mondo_continente1 = confronto_descrittori_statistici_casi_continenti_mondo(df_last_total_cases, continente1_casi, "World")
desc_stat_casi_mondo_continente2 = confronto_descrittori_statistici_casi_continenti_mondo(df_last_total_cases, continente2_casi, "World")

print(f"\nDescrittori statistici per il numero di casi nei 2 continenti scelti rispetto al mondo:")
print(f"{continente1_casi}:\n")
for chiave, valore in desc_stat_casi_mondo_continente1.items():
   print(f"{chiave}: {valore}")

# Aggiungi una linea vuota per separare gli output
print("\n" + "-"*50)

print(f"{continente2_casi}:\n")
for chiave, valore in desc_stat_casi_mondo_continente2.items():
   print(f"{chiave}: {valore}")
    
```

```python
# Calcola i descrittori statistici tra due continenti e il mondo per il numero di vaccinazioni
def confronto_descrittori_statistici_vaccinazioni_continenti_mondo(dataset, continente1, continente2):
   subset = dataset[(dataset["continent"] == continente1) | (dataset["continent"] == continente2)]

   total_vaccinations_sum = dataset[dataset["continent"].notna()]["total_vaccinations"].sum()
   perc_vaccinations = round((subset["total_vaccinations"].sum() / total_vaccinations_sum) * 100, 6) # Arrotonda la percentuale dopo la virgola
   
   # Calcola il minimo, il massimo, la media e la somma dei valori in total_cases, rispetto al mondo, formattati con i separatori per le cifre
   min = "{:,.2f}".format(subset["total_vaccinations"].min())
   max = "{:,.2f}".format(subset["total_vaccinations"].max())
   mean = "{:,.2f}".format(subset["total_vaccinations"].mean())
   sum = "{:,.2f}".format(subset["total_vaccinations"].sum())

   descr_statistici = {
      "Minimo": min,
      "Massimo": max,
      "Media": mean,
      "Somma": sum,
      "Percentuale (%)": perc_vaccinations 
   }

   return descr_statistici



```

```python
# Aggiungi una linea vuota per separare gli output
print("\n" + "-"*50)

continenti = ["Africa", "Asia", "Europe", "North America", "Oceania", "South America"]
print(f"\nDescrittori statistici per il numero di vaccinazioni nei continenti, rispetto al mondo:")

for continente in continenti:
    desc_stat_vaccinazioni_continente_mondo = confronto_descrittori_statistici_vaccinazioni_continenti_mondo(df_last_total_vaccinations, continente, "World")
    
    print(f"{continente}:\n")
    for chiave, valore in desc_stat_vaccinazioni_continente_mondo.items():
        print(f"{chiave}: {valore}")
    
    # Aggiungi una linea vuota per separare gli output
    print("\n" + "-"*50)
    
```
