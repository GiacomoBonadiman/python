```python
#Line plot
import numpy as np
import matplotlib.pyplot as plt

years = [2006 + x for x in range(16)]
weights = [80,83,84,85,86,82,81,79,83,80,82,82,83,81,80,79]

plt.scatter(years,weights)
plt.plot(years,weights, color="g", lw="5", linestyle="--")
#color and style "r--"
plt.show()
```

```python
#Bar plot
x=["C++","C#","Python","Java","Go"]
y=[20,50,140,1,45]
plt.bar(x,y, color="r", align="edge", width=0.1, edgecolor="green", lw=4)
plt.show()
```

```python
#Histogram plot
ages=np.random.normal(20,1.5,1000)
plt.hist(ages, bins=25, cumulative=True)
#bins=[ages.min(),18,21,ages.max()]
plt.show()
```

```python
#Pie charts
langs = ["Python", "C++", "Java", "C#", "Go"]
votes = [50,24,14,6,17]
explodes = [0,0,0.2,0,0]
plt.pie(votes, labels=langs, explode=explodes, autopct="%.2f%%", pctdistance=0.8, startangle=90)
plt.show()
```

```python
#Boxplot
heights=np.random.normal(172,8,300)
plt.boxplot(heights)
plt.show()

first = np.linspace(0,10,25)
second = np.linspace(10,200,25)
third = np.linspace(200,210,25)
fourth = np.linspace(210,230,25)

data = np.concatenate((first,second,third,fourth))
plt.boxplot(data)
plt.show()
```

```python
#Plot customization
import numpy as np
import matplotlib.pyplot as plt

years=[2014,2015,2016,2017,2018,2019,2020,2021]
income=[55,56,62,61,72,72,73,75]
income_ticks=list(range(50,81,2))

plt.plot(years,income)
plt.title("Income of John (in USD)", fontsize=30, fontname="Cambria")
plt.xlabel("Year")
plt.ylabel("Yearly income in USD")
plt.yticks(income_ticks,[f"${x}k" for x in income_ticks])
plt.show()
```

```python
#Legends and multiple plots
stock_a=[100,102,99,101,101,100,102]
stock_b=[90,95,102,104,105,105,109]
stock_c=[110,115,100,105,100,98,95]
plt.plot(stock_a,label="Company1")
plt.plot(stock_b,label="Company2")
plt.plot(stock_c,label="Company3")
plt.legend(loc="lower center")
plt.show()

votes=[10,2,5,16,22]
people=["A", "B", "C", "D", "E"]
plt.pie(votes, labels=None)
plt.legend(labels=people, loc="lower left")
plt.show()
```

```python
#Plot Styling
from matplotlib import style
style.use("dark_background") #Website: matplotlib.org

#Multiple Figures
x1,y1=np.random.random(100), np.random.random(100)
x2,y2=np.arange(100),np.random.random(100)

plt.figure(1)
plt.scatter(x1,y1)
plt.figure(2)
plt.plot(x2,y2)

#Subplots
x=np.arange(100)
fig,axs=plt.subplots(2,2)
axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title("Sine Wave")

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title("Cosine Wave")

axs[1,0].plot(x,np.random.random(100))
axs[1,0].set_title("Random Function")

np.seterr(divide = 'ignore') 
axs[1,1].plot(x,np.log(x))
axs[1,1].set_title("Log Function")
axs[1,1].set_xlabel("TEST")

fig.suptitle("Four Plots")
plt.show()
#plt.savefig("fourPlots.png", dpi=300, transparent=False, bbox_inches="Tight", pad_inches=20.2)
plt.tight_layout()
```

```python
#3D Plotting
import numpy as np
import matplotlib.pyplot as plt
ax = plt.axes(projection="3d")
x=np.random.random(100) #np.arange(0,50,0.1)
y=np.random.random(100) #np.sin(x) or np.arange(0,50,0.1)
z=np.random.random(100) #np.cos(x) or np.cos(x+y)
ax.scatter(x,y,z) #ax.plot(x,y,z)

ax.set_title("3D Plot")
ax.set_xlabel("test")
plt.show()
```

```python
a = np.arange(-5,5,0.1)
b = np.arange(-5,5,0.1)

A,B = np.meshgrid(a,b)
C = np.sin(A) * np.cos(B)
ax.plot_surface(A,B,C) #cmap="Spectral"
ax.view_init(azim=0,elev=90)

ax.set_title("3D Plot")
ax.set_xlabel("test2")
plt.show()
```

```python
#Animating Plots
import random
heads_tails = [0,0]
for _ in range(10):
    heads_tails[random.randint(0,1)] += 1
    plt.bar(["Heads","Tails"], heads_tails, color=["red","blue"])
    plt.pause(0.001)
plt.show()
```

```python
#SEABORN
import seaborn as sns
# Get the available Seaborn dataset names
sns.get_dataset_names()

# Load the datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets=sns.load_dataset("planets")

sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size", palette="YlGnBu")

```

```python
# Create plots with Seaborn
sns.histplot(tips['tip'], kde=True, bins=15)
sns.displot(tips['tip'], kde=True, bins=15)
sns.barplot(x="sex", y="tip", data=tips, palette="YlGnBu")
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex", palette="YlGnBu")
sns.stripplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu", dodge=True)
sns.jointplot(x="tip", y="total_bill", data=tips, kind="kde", fill=True, cmap="YlGnBu")
#kind="reg" or kind="hex" without shade
sns.pairplot(titanic.select_dtypes(["number"]),hue="pclass")
# Select numeric columns from the "titanic" dataset
numeric_columns_titanic = titanic.select_dtypes(include=['float64', 'int64'])
```

```python
# Create the correlation matrix for the "titanic" dataset
correlation_matrix = numeric_columns_titanic.corr()
```

```python
# Create a heatmap of the correlation matrix with Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm") #YlGnBu, icefire
```
